```python
#importing pandas, matplotlib, and seaborn
import pandas as pd

import matplotlib.pyplot as plt

import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
#importing the datasets

df_orders = pd.read_csv(r"C:\Users\DELL\Downloads\orders.csv")
df_order_details = pd.read_csv(r"C:\Users\DELL\Downloads\order_details.csv")
df_pizzas = pd.read_csv(r"C:\Users\DELL\Downloads\pizzas.csv")
df_pizza_types = pd.read_csv(r"C:\Users\DELL\Downloads\pizza_types.csv", encoding="latin1")
```

```python
#joining df_order_details and df_orders
df_orders_full = pd.merge(
    df_order_details,
    df_orders,
    on="order_id",
    how="inner")

```

```python
#joining df_orders_full and df_pizzas
df_orders_pizzas = pd.merge(
    df_orders_full,
    df_pizzas,
    on="pizza_id",
    how="inner")

```

```python
#joining all tables together

final_df = pd.merge(
    df_orders_pizzas,
    df_pizza_types,
    on="pizza_type_id",
    how="inner")

```

```python
# Extract the day of the week
final_df["day_of_week"] = final_df["date"].dt.day_name()
```

```python
# Find the total sales made on each day of the week. 
sales_by_day = (
    final_df
    .groupby("day_of_week")["revenue"]
    .sum()
    .sort_values(ascending=False))


sales_by_day
```

```python

days_order = [
    "Monday", "Tuesday", "Wednesday",
    "Thursday", "Friday", "Saturday", "Sunday"
]

sales_by_day = sales_by_day.reindex(days_order)
sales_by_day
```

## I calculated total sales by day of the week by extracting weekday names from order dates and aggregating revenue. The highest sales were recorded on [top_sales_day].

```python
# Which day of the week is when sales are made the most?
top_sales_day = sales_by_day.idxmax()
top_sales_value = sales_by_day.max()

top_sales_day, top_sales_value
```

```python
print(f"The Highest Sales Day of The Week Is: {top_sales_day:}, {top_sales_value}")
```

```python
# Extract month (year + month)
final_df["year_month"] = final_df["date"].dt.to_period("M")
```

```python
# Calculate sales per month

monthly_sales = (
    final_df
    .groupby("year_month")["revenue"]
    .sum()
    .sort_index())

monthly_sales
```

## Monthly sales show noticeable fluctuations, with peak sales occurring in months of July, May, March, and October, suggesting seasonality in customer demand

```python
monthly_sales_df = monthly_sales.reset_index()
monthly_sales_df["year_month"] = monthly_sales_df["year_month"].astype(str)
monthly_sales_df
```

```python
# Total quantity sold per pizza

pizza_performance = (
    final_df
    .groupby("name")["quantity"]
    .sum()
    .sort_values())


pizza_performance
```

```python
# Identify pizzas that are not doing well

low_performing_pizzas = pizza_performance.head(10)
low_performing_pizzas
```

```python
print(f"The Pizzas That Are Not Doing Well On The Menu are: {low_performing_pizzas}")
```

```python
# Identify pizzas that are not doing well be their revenue

low_revenue_pizzas = (
    final_df
    .groupby("name")["revenue"]
    .sum()
    .sort_values()
    .head(10))


low_revenue_pizzas
```

```python
# combine the pizzas that are not doing well by thier total_revenue and total_quantity
pizza_summary = (
    final_df
    .groupby("name")
    .agg(
        total_quantity=("quantity", "sum"),
        total_revenue=("revenue", "sum")
    )
    .sort_values("total_quantity"))

pizza_summary.head(10)
```

## Yes, there are pizza types on the menu that are underperforming. These pizzas consistently record low sales volume and revenue, indicating weak customer demand. This insight suggests opportunities for menu optimization, promotions, or possible removal.
